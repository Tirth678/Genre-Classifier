```python
from sklearn.tree import  DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import joblib
music_data = pd.read_csv('/Users/tirth/Developer/PythonTest/music.csv')
music_data.head()
# writing code for model
x = music_data.drop(columns=['genre']) # input data set
y = music_data['genre'] # output data set
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2) # we are allocating data for testing 20%
# music_data.info() 
model = DecisionTreeClassifier()
# model.fit(x,y)
model.fit(X_train, Y_train)
music_data
# model.predict([[21, 1], [22,0]]) # model will make prediction based on the input data set
# 21 input and 1 is gender
predictions = model.predict(X_test)
joblib.dump(model, 'music-recommender.joblib') # saving the model to disk
score = accuracy_score(Y_test, predictions)
score
print(f"Accuracy: {score * 100:.2f}%")
tree.export_graphviz(model, out_file='music-recommender.dot', feature_names=['age', 'gender'], class_names=sorted(y.unique()), label='all', rounded=True, filled=True)
```
